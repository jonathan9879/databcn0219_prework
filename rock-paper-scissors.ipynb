```python
# Import the choice function of the random module
# https://stackoverflow.com/questions/306400/how-to-randomly-select-an-item-from-a-list
from random import choice
# Assign to a list the 3 possible options: 'stone', 'paper' or 'scissors'.
hands = ['stone','paper','scissors']
# Assign a variable to the maximum number of games: 1, 3, 5, etc ...
maximum_games = 5
# Assign a variable to the number of games a player must win to win.
# Preferably the value will be based on the number of maximum games
games_to_win = (maximum_games / 2) + 1
# Define a function that randomly returns one of the 3 options.
# This will correspond to the play of the machine. Totally random.
def random_hand():
    return choice(hands)
# Define a function that asks your choice: 'stone', 'paper' or 'scissors'
# you should only allow one of the 3 options. This is defensive programming.
# If it is not stone, paper or scissors keep asking until it is.
def ask_user_hand():
    user_hand = None
    while user_hand not in hands:
        user_hand = input('What is your hand? Is it stone, is it paper or is it scissors?')
    return user_hand


# Define a function that resolves a combat.
# Returns 0 if there is a tie, 1 if the machine wins, 2 if the human player wins
def combat(machine_hand, human_hand):
    if machine_hand == human_hand: # Tie
        return 0
    elif machine_hand == 'paper':
        if human_hand == 'rock':
            return 1
        else:
            return 2
    elif machine_hand == 'rock':
        if human_hand == 'scissors':
            return 1
        else:
            return 2
    elif machine_hand == 'scissors':
        if human_hand == 'paper':
            return 1
        else:
            return 2
    
# Define a function that shows the choice of each player and the state of the game
# This function should be used every time accumulated points are updated
def state_of_game(machine_hand, human_hand, round_winner):
    print('The machine has chosen: ' + machine_hand)
    print('The human has chosen: ' + human_hand)
    
    if round_winner == 0:
        print('There has been a draw!\n')
    elif round_winner == 1:
        print('The machine has won this round!\n')
    else:
        print('The human has won this round!\n')
        
    
# Create two variables that accumulate the wins of each participant
machine_wins = 0
human_wins = 0

# Create a loop that iterates while no player reaches the minimum of wins
# necessary to win. Inside the loop solves the play of the
# machine and ask the player's. Compare them and update the value of the variables
# that accumulate the wins of each participant.
games_played = 0
while(games_played < maximum_games and machine_wins < games_to_win and human_wins < games_to_win):
    games_played += 1 # We count a new game and ask the machine and human for their hand of choice
    machine_hand = random_hand() 
    human_hand = ask_user_hand()
    
    round_winner = combat(machine_hand, human_hand) # We calculate who won and give the points to the winner
    if round_winner == 1:
        machine_wins += 1
    elif round_winner == 2:
        human_wins += 1
        
    state_of_game(machine_hand, human_hand, round_winner) # We output state of game

    
# Print by console the winner of the game based on who has more accumulated wins
if machine_wins > human_wins:
    print('The machine has won the game with a score of ' + str(machine_wins) + ' vs ' + str(human_wins))
elif human_wins > machine_wins:
    print('The human has won the game with a score of ' + str(human_wins) + ' vs ' + str(machine_wins))
else:
    print('There has been a draw! They both have a score of ' + str(machine_wins))
```
